```python
import pandas as pd
import numpy as np
from fbprophet import Prophet
df=pd.read_csv('C:\\Users\\user\\Desktop\\data\\07. example_wp_peyton_manning.csv ')

# 필요한 컬럼은 ds,y 2개이다.
#예측값의 상한과 하한을 제어해야 하면 , Cap, floor 컬럼에 값 지정
                                                                
df['cap'] = 6
df['floor'] = 1.5

m = Prophet() # Default growth='linear'
# prophet는 fit을 한번만 할수있다. 여러번 하려면 새로운 객체 생성 필요
m.fit(df)
```

```python
# 상항과 하한을 설정할 경우엔 prophet 객체를 생성할때 logistic 추가해야한다.
m = Prophet(growth='logistic')
m.fit(df)
```

```python
#미래 dataframe을 생성한다
future = m.make_future_dataframe(periods=365)
future.tail()
```

```python
#미래데이터에도 상한과 하한을 동일하게 지정해준다.
future['cap'] = 6
future['floor'] = 1.5
```

```python
# 예측
forecast = m.predict(future)
forecast.tail()
```

```python
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail(40)
```

```python
#시각화
fig1=m.plot(forecast)
```

```python
#forecast component 시각화 (Trend , weekly, yearly)
fig2 = m.plot_components(forecast)
```

# Trend Change Points

prophet에선 기본적으로 트렌드가 변경되는 지점을 자동으로 감지해서 트렌드 예측

하지만 사용자가 직접 조절가능
<br>
changepoint_range,<br>
changepoint_prior_scale,<br>
changepoints,

# changepoint_range


기본적으로 Prophet은 시계열 데이터의 80% 크기에서 잠재적으로 ChangePoint를 지정
90% 만큼 ChangePoint로 지정하고 싶다면 아래와 같이 설정


```python
m = Prophet(changepoint_range=0.9)
```

# changepoint_prior_scale

오버피팅이 심하면 너무 유연한 그래프가 나와서 모든 값에 근접하고 언더피팅일 경우 유연성이 부족하다<br>
기본값은 0.05이다 <br>
이 값을 늘리면 그래프가 유연해지고(=언더피팅 해결)<br>
이 값을 줄이면 유연성이 감소한다.(=오버피팅 해결)<br>

```python
m = Prophet(changepoint_prior_scale=0.05)
```

# changepoints(list)

잠재적으로 change point일 수 있는 날짜들 <br>
명시하지 않으면 잠재적인 changepoint가 자동으로 설정


```python
m = Prophet(changepoints=['2019-02-04', '2019-02-05'])
```

# Seasonality, Holiday Effects, And Regressors

## Model Holidays and Special Events
휴일이나 모델에 반영하고 싶은 이벤트 Dataframe으로 생성해서 반영
이벤트는 과거데이터와 미래 데이터가 모두 포함되어있어야한다.
주변 날짜를 포함시키기위해 lower_window, upper_window를 사용해서 업데이트의 영향을 조절가능하다.
예제는 Playoff경기일과 SuperBowl 경기날을 Holiday로 설정했다.


```python
playoffs = pd.DataFrame({
'holiday': 'playoff',
'ds': pd.to_datetime(['2008-01-13', '2009-01-03', '2010-01-16',
                      '2010-01-24', '2010-02-07', '2011-01-08',
                      '2013-01-12', '2014-01-12', '2014-01-19',
                      '2014-02-02', '2015-01-11', '2016-01-17',
                      '2016-01-24', '2016-02-07']),
'lower_window': 0,
'upper_window': 1,
})
superbowls = pd.DataFrame({
    'holiday': 'superbowl',
    'ds': pd.to_datetime(['2010-02-07', '2014-02-02', '2016-02-07']),
    'lower_window': 0,
    'upper_window': 1,
  })
holidays = pd.concat((playoffs, superbowls))
```

사용하는 방법은 간단하다.
Prophet객체를 생성할때 Holiday 인자로 넘기면 된다.

```python
m = Prophet(holidays=holidays)
forecast = m.fit(df).predict(future)
```

Holiday Effect를 아래 코드로 확인할 수 있다.

```python
forecast[(forecast['playoff'] + forecast['superbowl']).abs() > 0][
      ['ds', 'playoff', 'superbowl']][-10:]
```

plot_component로 시각화할 경우 holidays의 영향도 볼 수 있음


```python
fig = m.plot_components(forecast)
```

만약에 holiday에 오버피팅된것 같으면 holidays_prior_scale을 조정해서 smooth하게 변경이 가능하다.

```python
m = Prophet(holidays=holidays, holidays_prior_scale=0.05).fit(df)
forecast = m.predict(future)

forecast[(forecast['playoff'] + forecast['superbowl']).abs() > 0][
      ['ds', 'playoff', 'superbowl']][-10:]
```

# Fourier Order for Seasonalities

Seaonsalities 부분 푸리에의 합을 사용하여 추정한다.<br>
푸리에 급수는 주기함수를 삼각함수의 급수로 나타낸것이다.<br>
yearly_seasonality 파라미터는 10으로 디폴트.
만약 시즈널리티가 자주발생한다고하면 이 값을 20으로 수정하자 ! <br>
다만 오버피팅을 조심하자.<br>

```python
from fbprophet.plot import plot_yearly

m = Prophet(yearly_seasonality=10).fit(df)
a = plot_yearly(m)
```

# Specify Custom Seasonalities

커스텀 시즈널리티를 생성할 수 있음.<br>
기본적으로 Weekly, Yearly 특성을 사용한다.<br>
m.add_seasonality로 추가하며 name, period, fourier_order가 있다.


```python
m = Prophet(weekly_seasonality=False)
m.add_seasonality(name='monthly', period=30.5, fourier_order=5)
forecast = m.fit(df).predict(future)
fig = m.plot_components(forecast)
```

# Uncertainly Intervals

불확실성의 범위가 나타나는 원인<br>
1) Trend 불확실성<br>
2) Seasonality 추정의 불확실성<br>
3) 추가 관찰되는 잡음<br>
<br><br>
Uncertainty in the Trend <br>
예측을 하면 yhat_lower, yhat_upper가 나타나는데 이 범위도 사용자가 조절할 수 있다. <br>
interval_width 의 기본 값은 80%
<br> changepoint_prior_scale을 조절하면 예측 불확실성이 증가한다.<br>

```python
forecast = Prophet(interval_width=0.95).fit(df).predict(future)
```

Uncertainty in seasonality <br>
베이지안 샘플링을 사용해야한다.<br>
mcmc.samples 파라미터를 사용한다. <br>
이 값을 사용하면 최초 n일에 대해 적용한다는 뜻이다.<br>

```python
m = Prophet(mcmc_samples=300)
forecast = m.fit(df).predict(future)
fig = m.plot_components(forecast)
```
