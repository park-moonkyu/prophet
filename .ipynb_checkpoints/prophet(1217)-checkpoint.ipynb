```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.read_csv('./final.csv')
df=df[['DT','SALE_SUM_AMT']]
plot=df[['SALE_SUM_AMT']]
```

```python
df.rename(columns={'DT': 'ds', 'SALE_SUM_AMT': 'y'}, inplace=True)
df
```

```python
def modify_date(row):
    return row[:10]
```

```python
df['ds'] = df['ds'].apply(modify_date)
df.head()
```

```python
fig, axe = plt.subplots(nrows=1, figsize=(20,10))
sns.lineplot(x='ds',y='y' ,data=df,marker='8',linewidth = 3)

```

```python
df['y']=np.log(df['y'])
print(np.power(np.e,df['y'][0]))
```
