# 전체로 진행

Prophet은  sklearn 모델의 API를 따릅니다. 우리는 Prophet 클래스의 인스턴스를 만들고 그에 대한 fit과 predict methods를 불러옵니다.

Prophet 의 입력값(input)은 언제나 두개(ds 와 y)의 columns을 가진 데이터프레임(dataframe)입니다.
ds (datastamp)는 date 나 datetime(둘 중 어느것이어도 괜찮음) 을 반드시 포함해야하고, y column은 반드시 numeric하고 우리가 예측하고자하는 값을 나타내어야 합니다.

```python
import pymssql
import pandas as pd
import numpy as np
from fbprophet import Prophet

# MSSQL 접속
conn = pymssql.connect(host=r"211.233.21.82", user='bif', password='DWBIF2020!@#', database='DWBIF', charset='cp949')

# Connection 으로부터 Cursor 생성
cursor = conn.cursor()
 
# SQL문 실행
cursor.execute('SELECT * FROM DWBIF.dbo.V_TBIF1010_VALUABLE ORDER BY YY, WK;')
row = cursor.fetchall()
cols = [item[0] for item in cursor.description]
df=pd.DataFrame(row,columns=cols)


# 필요한 컬럼은 ds,y 2개이다.
#예측값의 상한과 하한을 제어해야 하면 , Cap, floor 컬럼에 값 지정
                         
df=df[['DT','SALE_SUM_AMT']]
df.rename(columns={'DT':'ds','SALE_SUM_AMT':'y'},inplace=True)
df.head()


# cap / floor 구분에 따른 maximium/ minimiun 설정.

"""
df['cap'] = 6
df['floor'] = 1.5
"""
```

```python
# 상항과 하한을 설정할 경우엔 prophet 객체를 생성할때 logistic 추가해야한다.

m = Prophet()
m.fit(df)
```

```python
#미래 dataframe을 생성한다
future = m.make_future_dataframe(periods=365)
future.tail()
```

```python

forecast = m.predict(future)

df= forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']]

df.to_csv('C:\\Users\\user\\Desktop\\result.csv')
```

```python
m.plot(forecast);
```


만약 예측 부분을 보고싶다면,
Prophet.plot_component 메소드를 사용하면 된다.
기본값으로 연간 계절성과 주간 계절성 시계열 트렌드를 보게될것이다.
공휴일을 포함하면 그것들도 볼 수 있다.


```python
m.plot_components(forecast);
```

# Trend Change Points

prophet에선 기본적으로 트렌드가 변경되는 지점을 자동으로 감지해서 트렌드 예측

하지만 사용자가 직접 조절가능
<br>
changepoint_range,<br>
changepoint_prior_scale,<br>
changepoints,

# changepoint_range


기본적으로 Prophet은 시계열 데이터의 80% 크기에서 잠재적으로 ChangePoint를 지정
90% 만큼 ChangePoint로 지정하고 싶다면 아래와 같이 설정


m = Prophet(changepoint_range=0.9)

# changepoint_prior_scale
오버피팅이 심하면 너무 유연한 그래프가 나와서 모든 값에 근접하고 언더피팅일 경우 유연성이 부족하다
기본값은 0.05이다
이 값을 늘리면 그래프가 유연해지고(=언더피팅 해결)
이 값을 줄이면 유연성이 감소한다.(=오버피팅 해결)

1
m = Prophet(changepoint_prior_scale=0.05)

# changepoints(list)
잠재적으로 change point일 수 있는 날짜들
명시하지 않으면 잠재적인 changepoint가 자동으로 설정

m = Prophet(changepoints=['2019-02-04', '2019-02-05'])
