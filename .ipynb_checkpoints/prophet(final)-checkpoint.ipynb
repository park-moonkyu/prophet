```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import fbprophet as ph
import datetime

df1 = pd.read_csv('./final.csv')
df=df1[['DT','SALE_SUM_AMT']]
plot=df[['SALE_SUM_AMT']]

train=df[:-53]
test=df[-52:]

# df를 train으로 설정
df=train
```

```python
df.rename(columns={'DT': 'ds', 'SALE_SUM_AMT': 'y'}, inplace=True)
df
```

```python
def modify_date(row):
    return row[:10]
```

```python
df1.rename(columns={'DT': 'ds', 'SALE_SUM_AMT': 'y'}, inplace=True)
df1['ds'] = df1['ds'].apply(modify_date)
print(df1['y'][0])
df1['y']=np.log(df1['y'])

print(df1['y'][0])
df1.to_csv('./real.csv')
fig, axe = plt.subplots(nrows=1, figsize=(20,10))
sns.lineplot(x='ds',y='y' ,data=df1,marker='8',linewidth = 3)

```

```python
df['ds'] = df['ds'].apply(modify_date)
df.head()
```

```python
fig, axe = plt.subplots(nrows=1, figsize=(20,10))
sns.lineplot(x='ds',y='y' ,data=df,marker='8',linewidth = 3)

```

```python
df['y']=np.log(df['y'])
print(np.power(np.e,df['y'][0]))
print(df['y'][0])
```

```python
df['ds'] = pd.to_datetime(df['ds'])
df.dtypes
```

```python
m= ph.Prophet()
m.fit(df)
```

```python
m.history
```

```python
future = m.make_future_dataframe(freq='W',periods=52)
```

```python
forecast= m.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
m.plot(forecast)
```

```python
m.plot_components(forecast)
```

```python
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].to_csv('./scale(0.5).csv')
```

# Automatic changepoint detection in Prophet (Finding better scale)

If the trend changes are being overfit (too much flexibility) or underfit (not enough flexiblity), you can adjust the strength of the sparse prior using the input argument changepoint_prior_scale.



```python
future_ch = m.make_future_dataframe(freq='W', periods=52) # Next 31 days
m_ch = ph.Prophet(changepoint_prior_scale=1) # default is 0.5
forecast_ch = m_ch.fit(df).predict(future_ch)
m_ch.plot(forecast_ch);
```

```python
forecast_ch= m_ch.predict(future_ch)
forecast_ch[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
forecast_ch[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].to_csv('./scale(1).csv')
```
